```python
import pandas as pd
import numpy as np
```

```python
acad = pd.read_csv("academic_awards.csv")
```

```python
acad.iloc[0:106]
```

```python
acad["rank"] = np.zeros(len(acad)).astype(int)
acad
```

```python
for k in range(0,len(acad)):
    if(acad["Monsoon 21"][k] == 'Deans List 1'):
        acad["rank"][k] += 4;
    elif(acad["Monsoon 21"][k] == 'Deans List 2'):
        acad["rank"][k] +=3;
    elif(acad["Monsoon 21"][k] == 'Deans List 3'):
        acad["rank"][k] +=2;
    elif(acad["Monsoon 21"][k] == 'Merit List'):
        acad["rank"][k] +=1;
        
for k in range(0,len(acad)):       
    if(acad["Spring 22"][k] == 'Deans List 1'):
        acad["rank"][k] += 4;
    elif(acad["Spring 22"][k] == 'Deans List 2'):
        acad["rank"][k] +=3;
    elif(acad["Spring 22"][k] == 'Deans List 3'):
        acad["rank"][k] +=2;
    elif(acad["Spring 22"][k] == 'Merit List'):
        acad["rank"][k] +=1;
```

```python
acad.sort_values(by=["rank"],ascending=[False],inplace=True)
```

```python
acad
```

```python
acad.reset_index(drop=True,inplace=True)
```

```python
acad["S.No"] = pd.Series(range(1,107,1))
```

```python
acad.to_csv("acad_sorted.",index=False)
```
